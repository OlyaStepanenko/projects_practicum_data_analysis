**Функция для расчета границ нормального распределения**

```python
# функция для расчета границ нормального распределения
def normal_borders(df, column):
    left_border=df[column].quantile(0.25)-1.5*(df[column].quantile(0.75)-df[column].quantile(0.25))
    right_border=df[column].quantile(0.75)+1.5*(df[column].quantile(0.75)-df[column].quantile(0.25))
    if left_border < 0:
        left_border = 0
    return left_border, right_border
```

**Функция для расчета удержания**

```python
def get_retention(
    profiles,
    sessions,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    # добавляем столбец payer в передаваемый dimensions список
    dimensions = ['event_target'] + dimensions

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # собираем «сырые» данные для расчёта удержания
    result_raw = result_raw.merge(
        sessions[['user_id', 'event_time']], on='user_id', how='left'
    )
    result_raw['lifetime'] = (
        result_raw['event_time'] - result_raw['first_ts']
    ).dt.days

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        result = result.div(result['cohort_size'], axis=0)
        result = result[['cohort_size'] + list(range(horizon_days))]
        result['cohort_size'] = cohort_sizes
        return result

    # получаем таблицу удержания
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # получаем таблицу динамики удержания
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )

    # возвращаем обе таблицы и сырые данные
    return result_raw, result_grouped, result_in_time
```

**Функция "сглаживания" графика**

```python
def filter_data(df, window):
    # для каждого столбца применяем скользящее среднее
    for column in df.columns.values:
        df[column] = df[column].rolling(window).mean() 
    return df
```

**Функция для визуализации удержания**

```python
def plot_retention(retention, retention_history, horizon, window=7):

    # задаём размер сетки для графиков
    plt.figure(figsize=(15, 10))

    # исключаем размеры когорт и удержание первого дня
    retention = retention.drop(columns=['cohort_size', 0])
    # в таблице динамики оставляем только нужный лайфтайм
    retention_history = retention_history.drop(columns=['cohort_size'])[
        [horizon - 1]
    ]

    # если в индексах таблицы удержания только payer,
    # добавляем второй признак — cohort
    if retention.index.nlevels == 1:
        retention['cohort'] = 'All users'
        retention = retention.reset_index().set_index(['cohort', 'event_target'])

    # в таблице графиков — два столбца и две строки, четыре ячейки
    # в первой строим кривые удержания платящих пользователей
    ax1 = plt.subplot(2, 2, 1)
    retention.query('event_target == True').droplevel('event_target').T.plot(
        grid=True, ax=ax1
    )
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание совершивших\nцелевое действие пользователей')

    # во второй ячейке строим кривые удержания неплатящих
    # вертикальная ось — от графика из первой ячейки
    ax2 = plt.subplot(2, 2, 2, sharey=ax1)
    retention.query('event_target == False').droplevel('event_target').T.plot(
        grid=True, ax=ax2
    )
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание не совершивших\nцелевой действие пользователей')

    # в третьей ячейке — динамика удержания платящих
    ax3 = plt.subplot(2, 2, 3)
    # получаем названия столбцов для сводной таблицы
    columns = [
        name
        for name in retention_history.index.names
        if name not in ['dt', 'event_target']
    ]
    # фильтруем данные и строим график
    filtered_data = retention_history.query('event_target == True').pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax3)
    plt.xlabel('Дата привлечения')
    plt.title(
        'Динамика удержания\nсовершивших целевое действие\nпользователей  на {}-й день'.format(
            horizon
        )
    )

    # в чётвертой ячейке — динамика удержания неплатящих
    ax4 = plt.subplot(2, 2, 4, sharey=ax3)
    # фильтруем данные и строим график
    filtered_data = retention_history.query('event_target == False').pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax4)
    plt.xlabel('Дата привлечения')
    plt.title(
        'Динамика удержания\nне совершивших целевое действие\nпользователей на {}-й день'.format(
            horizon
        )
    )
    
    plt.tight_layout()
    plt.show()
```

**Функция для расчёта конверсии**

```python
# функция для расчёта конверсии

def get_conversion(
    profiles,
    purchases,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # определяем дату и время первой покупки для каждого пользователя
    first_purchases = (
        purchases.sort_values(by=['user_id', 'event_date'])
        .groupby('user_id')
        .agg({'event_date': 'first'})
        .reset_index()
    )

    # добавляем данные о покупках в профили
    result_raw = result_raw.merge(
        first_purchases[['user_id', 'event_date']], on='user_id', how='left'
    )

    # рассчитываем лайфтайм для каждой покупки
    result_raw['lifetime'] = (
        result_raw['event_date'] - result_raw['first_ts']
    ).dt.days

    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users' 
        dimensions = dimensions + ['cohort']

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        result = result.fillna(0).cumsum(axis = 1)
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # делим каждую «ячейку» в строке на размер когорты
        # и получаем conversion rate
        result = result.div(result['cohort_size'], axis=0)
        result = result[['cohort_size'] + list(range(horizon_days))]
        result['cohort_size'] = cohort_sizes
        return result

    # получаем таблицу конверсии
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # для таблицы динамики конверсии убираем 'cohort' из dimensions
    if 'cohort' in dimensions: 
        dimensions = []

    # получаем таблицу динамики конверсии
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )

    # возвращаем обе таблицы и сырые данные
    return result_raw, result_grouped, result_in_time
```

**Функция для визуализации конверсии**

```python
def plot_conversion(conversion, conversion_history, horizon, window=7):

    # задаём размер сетки для графиков
    plt.figure(figsize=(15, 5))

    # исключаем размеры когорт
    conversion = conversion.drop(columns=['cohort_size'])
    # в таблице динамики оставляем только нужный лайфтайм
    conversion_history = conversion_history.drop(columns=['cohort_size'])[
        [horizon - 1]
    ]

    # первый график — кривые конверсии
    ax1 = plt.subplot(1, 2, 1)
    conversion.T.plot(grid=True, ax=ax1)
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Конверсия пользователей')

    # второй график — динамика конверсии
    ax2 = plt.subplot(1, 2, 2, sharey=ax1)
    columns = [
        # столбцами сводной таблицы станут все столбцы индекса, кроме даты
        name for name in conversion_history.index.names if name not in ['dt']
    ]
    filtered_data = conversion_history.pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax2)
    plt.xlabel('Дата привлечения')
    plt.title('Динамика конверсии пользователей на {}-й день'.format(horizon))

    plt.tight_layout()
    plt.show()
```

**Функция для проверки гипотез z-тестом**

```python
def check_hypothesise(successes1, successes2, trials1, trials2):
    p1 = successes1 / trials1
    p2 = successes2 / trials2
    p_combined = (successes1 + successes2) / (trials1 + trials2)
    difference = p1 - p2
    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials1 + 1/trials2))
    distr = st.norm(0, 1)
    p_value = (1 - distr.cdf(abs(z_value))) * 2
    return p_value
```

**Разделим события на сессии.**

В логе событий нет информации о начале и конце пользовательской сессии, чтобы разделить события на сессии сначала посмотрим сколько времени пользователи проводят в приложении в течении дня.

```python
avg_time_per_day = (
    df_events.sort_values(by=['event_time'])
    .groupby(['event_date', 'user_id'], sort=False)
    .agg({'event_time':['first', 'last']})
    .droplevel(1, axis=1)
    .reset_index()
)
avg_time_per_day.columns = ['date', 'user_id', 'first_time', 'last_time']
#avg_time_per_day.head()
avg_time_per_day['time_minutes'] =(
    (avg_time_per_day['last_time'] - avg_time_per_day['first_time'])
    .astype('timedelta64[m]')
)
avg_time_per_day.head()
```

```python
print(
    f'Минимальное время, которое пользователи проводят в приложении в день',
    avg_time_per_day['time_minutes'].min() 
)
print(
    '\nМаксимальное время, проведенное пользователем в приложении в день',
    avg_time_per_day['time_minutes'].max() 
)
print(
    '\nСреднее время, которое пользователи проводят в приложении в день',
    avg_time_per_day['time_minutes'].mean() 
)
print(
    ' \nМедианное время, которое пользователи проводят в приложении в день',
    avg_time_per_day['time_minutes'].median() 
)
```

Как видно среднее значение в 7 раз больше медианного из-за пользователей, которые проводят в приложении значительно больше времени, чем большинство.

**События пользователей, совершивших и не совершивших целевое действие.**

```python
users_target = profiles.query('event_target == True')['user_id']

event_by_users_target = (
    df_events.query('user_id in @users_target and event_name != "contacts_show" and event_name != "contacts_call"')
    .groupby('event_name')
    .agg({'event_time' : 'count'})
    .reset_index()
    .rename(columns={'event_time' : 'total_events'})
)

event_by_users_no_target = (
    df_events.query('user_id not in @users_target')
    .groupby('event_name')
    .agg({'event_time' : 'count'})
    .reset_index()
)

events_by_users =(
    event_by_users_target.merge(event_by_users_no_target, on='event_name', how='left' )
    .rename(columns={'total_events': 'target', 'event_time' : 'no_target'})
    .set_index('event_name')
)
events_by_users['%_in_target'] = (
    round(events_by_users['target'] / events_by_users['target'].sum() * 100, 2)
)
events_by_users['%_in_no_target'] = (
    round(events_by_users['no_target'] / events_by_users['no_target'].sum() * 100, 2)
)

(events_by_users[['%_in_no_target', '%_in_target']]
 .sort_values(by='%_in_target')
 .plot(kind='barh',figsize=(10, 5), grid=True)
)
plt.title('События пользователей совершивших и не совершивших целевое действие.')
plt.legend(['События пользователей без целевого действия', 'События пользователей с целевым действием'])
plt.ylabel('Событие')
plt.xlabel('Доля событий')
plt.tight_layout() 
plt.show() 

events_by_users.sort_values(by='%_in_target', ascending=False)
```

```python
list_tips_show = df_events.query('event_name == "tips_show"')['user_id'].unique()
sessions_tips_show = (
    df_events.query('user_id in @list_tips_show')
    .groupby('event_name')
    .agg({'event_time' : 'count', 'user_id' :'nunique'})
    .reset_index()
    .rename(columns={'event_time' : 'events_with_tips_show', 'user_id' : 'users_tips_show'})
)

sessions_no_tips_show = (
    df_events.query('user_id not in @list_tips_show')
    .groupby('event_name')
    .agg({'event_time' : 'count', 'user_id' :'nunique'})
    .reset_index()
    .rename(columns={'event_time' : 'events_no_tips_show', 'user_id' : 'users_no_tips_show'})
)

sessions_on_tips_show =(
    sessions_tips_show.merge(sessions_no_tips_show, on='event_name', how='outer' )
    .set_index('event_name')
)
sessions_on_tips_show['%_in_events_tips'] = (
    round(sessions_on_tips_show['events_with_tips_show'] 
          / sessions_on_tips_show['events_with_tips_show'].sum() * 100, 2)
)
sessions_on_tips_show['%_in_events_no_tips'] = (
    round(sessions_on_tips_show['events_no_tips_show'] /
          sessions_on_tips_show['events_no_tips_show'].sum() * 100, 2)
)

sessions_on_tips_show['%_in_users_tips'] = (
    round(sessions_on_tips_show['users_tips_show'] 
          / len(list_tips_show) * 100, 2)
)

sessions_on_tips_show['%_in_users_no_tips'] = (
    round(sessions_on_tips_show['users_no_tips_show'] 
          / (df_events.query('user_id not in @list_tips_show')['user_id'].nunique()) * 100, 2)
)

print(
    'Всего сессий с событием "tips_show" - {}'
    .format(df_events.query('user_id in @list_tips_show')['session_id'].nunique()        
    )
)

print(
    'Всего сессий без события "tips_show" - {}'
    .format(df_events.query('user_id not in @list_tips_show')['session_id'].nunique()        
    )
)

(sessions_on_tips_show[['%_in_users_tips', '%_in_users_no_tips']]
 .sort_values(by='%_in_users_tips')
 .plot(kind='barh',figsize=(10, 5), grid=True)
)
plt.title('События пользователей, в зависимости от показа рекомендованных объявлений.')
plt.legend(['Показали рекомендованные объявления', 'Не показали рекомендованные объявления'])
plt.ylabel('Событие')
plt.xlabel('Доля пользователей, совершивших событие')
plt.tight_layout() 
plt.show()

print ('\nСобытия пользователей, которым показали рекомендации')

display(
    sessions_on_tips_show[['events_with_tips_show', '%_in_events_tips', 'users_tips_show', '%_in_users_tips']]
    .sort_values(by='%_in_users_tips', ascending=False)
)

print ('\nСобытия пользователей, которым не показали рекомендации')

(
    sessions_on_tips_show[['events_no_tips_show', '%_in_events_no_tips', 'users_no_tips_show', '%_in_users_no_tips']] 
    .sort_values(by='%_in_users_no_tips', ascending=False)
)
```

События сессий этих двух групп отличаются. Пользователи, которым показаны рекомендации практически не смотрят фотографии (всего 0.5% пользователей), из действий предпочитают просмотр карты объявлений. Пользователи, которым не показали рекомендации в основном смотрят фотографии объявлений (72.5%) и осуществляют события поиска (58%). Также можно заметить, что у пользователей смотревших рекомендации доля совершивших целевое событие меньше (18%, у не смотревших - 31%).

Интересно, что при отсутствии показа рекомендаций, у второй группы пользователей есть событие `tips_click` (1.7% пользователей делали этот шаг).

Обычно рекомендованные объявления показывают всем пользователям по-умолчанию. В нашем случае могут быть разные причины отсутствия рекомендаций в сессиях. Либо это техническая проблема и в таблицы для исследования по какой-то причине не выгрузились события `tips_show` для второй группы пользователей.

## Проверка гипотез

### Гипотеза: Конверсии в целевое событие групп совершивших событие `tips_show` и `tips_click` и только `tips_show` различаются.

Одни пользователи совершают действия `tips_show` и `tips_click` , другие — только `tips_show` . Проверим гипотезу: конверсия в просмотры контактов различается у этих двух групп.

Предварительно посмотрим конверсию пользователей в зависимости от совершения ими шагов `tips_show` и `tips_click`

```python
conversion_raw, conversion_grouped, conversion_history = get_conversion(
    profiles, df_target, observation_date, horizon_days, dimensions=['event_tips']
)

plot_conversion(conversion_grouped, conversion_history, horizon_days)
```

Средняя конверсия 7-го дня у пользователей совершивших действия `tips_show` и `tips_click` значительно выше чем у пользователей совершивших только `tips_show` (27 и 16% соответственно). Если смотреть на показатели  в динамике, то за период исследования конверсия пользователей с действием `tips_show` остается примерно на одном уровне. Конверсия пользователей с действиями `tips_show` и `tips_click` не стабильна. На графики видно, ее резкое падение для когорт до 21 октября (для когорт 17-21 октября показатели конверсии сравниваемых групп примерно на одном уровне). Затем конверсия `tips_show` и `tips_click` начинает расти и у последних когорт достигает 35%.

Также можно отметить, что более высокие показатели конверсии за весь период у пользователей, которые вообще не совершали этих действий.

**Статистическая проверка гипотезы**

Посчитаем количество пользователей и пользователей, совершивших целевое событие для каждой группы

```python
tips_show = (
    df_target.loc[df_target['user_id']
                  .isin(profiles.query('event_tips == "tips_show"')['user_id'].unique())]
    .groupby('event_name')
    .agg({'user_id':'nunique'})
    .rename(columns={'user_id':'successes1'})
)

tips_show_and_click = (
    df_target.loc[df_target['user_id']
                  .isin(profiles.query('event_tips == "tips_show_and_click"')['user_id'].unique())]
    .groupby('event_name')
    .agg({'user_id':'nunique'})
    .rename(columns={'user_id':'successes2'})
)

tips_show['trials1'] = profiles.query('event_tips == "tips_show"')['user_id'].nunique()
tips_show_and_click['trials2'] = profiles.query('event_tips == "tips_show_and_click"')['user_id'].nunique()    
print('Группа "tips_show"')
display(tips_show)
print('\nГруппа "tips_show_and_click"')
tips_show_and_click
```

**Сформулируем нулевую и альтернативную гипотезы**

**Нулевая гипотеза** - Конверсии в целевое событие двух генеральных совокупностей пользователeй совершаюoщих действия `tips_show` и `tips_click` и пользователей совершающих только `tips_show` равны.

**Альтернативная гипотеза** - Конверсии в целевое событие двух генеральных совокупностей пользователeй совершаюoщих действия `tips_show` и `tips_click` и пользователей совершающих только `tips_show` различаются.

Определим уровень статистической значимости `alpha` для сравнения двух групп **0,01**
Применим функцию для расчета статистической значимости.

```python
alpha = 0.01

p_value = check_hypothesise(
    tips_show.loc['contacts_show','successes1'],
    tips_show_and_click.loc['contacts_show','successes2'],
    tips_show.loc['contacts_show','trials1'],
    tips_show_and_click.loc['contacts_show','trials2']
)
if p_value < alpha:
    print(
        f'Отвергаем нулевую гипотезу: \nмежду долями по целевому событию в группах есть значимая разница\n',
        'p_value=', p_value,3
    )
else:
    print(                     
        f'\nНе получилось отвергнуть нулевую гипотезу,\nнет оснований считать доли групп по целевому событию разными\n',
        'p_value=', p_value, 3,'\n'
    )
```

**Вывод**

Вероятность того, что различий в конверсиях между группами `tips_show` и  `tips_show_and_click` нет почти нулевая. Значит у нас есть основания отвергнуть нулевую гипотезу и не отвергать альтернативную.

**Альтернативная гипотеза** - Конверсии в целевое событие двух генеральных совокупностей пользователeй совершаюoщих действие `photos_show` и пользователей не совершающих его различаются.
