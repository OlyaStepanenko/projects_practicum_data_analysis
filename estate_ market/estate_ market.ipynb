Можно предположить, что количество комнат 0 соответствует однокомнатным квартирам студиям

```python
data.query('rooms == 0')['studio'].value_counts()
```

```python
data['studio'].value_counts()
```

```python
data.query('rooms != 0 and studio == True')
```

```python
data.loc[(data['rooms'] == 0) & (data['studio'] == True),'rooms'] = 1

print ('Количество объявлений с квартирами где 0 комнат:', len(data.query('rooms == 0')))
```

По всем объявлениям известна общая площадь. Посчитаем отношение жилой площади к общей в зависимости от количества комнат. И на основе этого показателя заполним пропущенные значения.

В квартирах-студиях и квартирах с открытой планировкой площадь кухни заполнять не будем, т.к. фактически в таких квартирах нет отдельной кухни.
В остальных квартирах незаполненные значения в столбце `kitchen_area` заполним аналогично столбцу `living_area` с помощью медианного показателя отношения площади кухни к общей площади квартиры в зависимости от количества комнат. В итоге в столбце `kitchen_area` должно остаться 216 пропущенных значений, соответствующих объявлениям с квартрирами-студиями и квартирами с открытой планировкой.

```python
print('Количество пропусков до:', data['kitchen_area'].isna().sum())
for room in data['rooms'].unique():
    kitchen_area_ratio = (
        (data.loc[(data['rooms'] == room) & (data['open_plan'] == False) & (data['studio'] == False) , 'kitchen_area']
         / data.loc[(data['rooms'] == room) & (data['open_plan'] == False) & (data['studio'] == False), 'total_area'])
        .median())
    print(room, kitchen_area_ratio)
    data.loc[(data['kitchen_area'].isna()) & (data['rooms'] == room) & (data['open_plan'] == False) & (data['studio'] == False), 'kitchen_area'] = (
        data.loc[(data['kitchen_area'].isna()) & (data['rooms'] == room) & (data['open_plan'] == False) & (data['studio'] == False), 'total_area']
        * kitchen_area_ratio)
    
print('Количество пропусков после:', data['kitchen_area'].isna().sum())
```

Чтобы построить читаемую гистограмму, ее границы лучше определить в пределах нормальных значений, исключая выбросы.

Посчитаем границы нормального распределения.
Напишем функцию для их расчета.

```python
def normal_borders(df, column):
    left_border=df[column].quantile(0.25)-1.5*(df[column].quantile(0.75)-df[column].quantile(0.25))
    right_border=df[column].quantile(0.75)+1.5*(df[column].quantile(0.75)-df[column].quantile(0.25))
    if left_border < 0:
        left_border = 0
    return left_border, right_border
```

```python
normal_borders(data, 'total_area') # применим функцию к столбцу 'total_area'
```

```python
data['total_area'].hist(range=(0, 115), bins=35, figsize=(15,5),)
plt.title('График распределения объявлений по общей площади квартир (в нормальных пределах до 115 кв м)')
plt.ylabel('количество объявлений')
plt.xlabel('общая площадь кв м')
```
